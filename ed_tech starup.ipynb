# Preparing Data for Segment Extraction

```python
arange_final = df_mod.sort_values(by='OVERALL_LI',ascending=False).head(15)
arange_final
```

```python
arange_final_b = df_mod.sort_values(by='OVERALL_LI',ascending=True).head(15)
arange_final_b
```

```python
final_states = pd.concat([arange_final,arange_final_b],axis=0,sort=False)
final_states
```

```python
final_states['P_RUR_POP']=100-final_states['P_URB_POP']
final_states['P_RUR_POP']
```

```python
df1 = final_states.drop(['STATNAME','STATCD','FEMALE_LIT','MALE_LIT','SEXRATIO','DISTRICTS','TOTPOPULAT','AREA_SQKM','SCHTOT','SCHTOTG','SCHTOTGR','SCHTOTPR','TCHTOTG', 'TCHTOTP', 'SCLSTOT', 'STCHTOT',  'SPLAYTOT','ENRTOT','ENRTOTG','ENRTOTGR','ENRTOTPR'],axis='columns')
df1
```

```python
def roundValues(x):
    df1[x] = np.round_(df1[x]/10)*10
    return df1

school = ['OVERALL_LI','P_URB_POP','P_RUR_POP']

for i in school:
    roundValues(i)
df1
```

```python
def roundthousand(x):
    df1[x] = np.round_(df1[x]/20000)*20000
    return df1

facility = ['ROADTOT','SWATTOT','SELETOT']

for i in facility:
    roundthousand(i)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
def labelling(x):
    df1[x] = LabelEncoder().fit_transform(df1[x])
    return df1

book = ['OVERALL_LI','P_URB_POP','P_RUR_POP','ROADTOT','SWATTOT','SELETOT']

for i in book:
    labelling(i)
df1
```

```python
book = ['OVERALL_LI','P_URB_POP','P_RUR_POP','ROADTOT','SWATTOT','SELETOT']
df1_factors = df1.loc[:,book]
df1_factors
```

```python
x = df1.loc[:,book].values
x
```

# Segment Extraction

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(x)

pca = PCA(n_components=6)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6']
pf = pd.DataFrame(data=pc,columns = names)
pf
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df1 = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df1['variable'] = df1_factors.columns.values
loadings_df1 = loadings_df1.set_index('variable')
loadings_df1
```

```python
plt.rcParams['figure.figsize'] = (10,10)
ax = sns.heatmap(loadings_df1, annot=True, cmap='Spectral')
plt.show()
```

```python
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
# get PC scores
pca_scores = PCA().fit_transform(x)

# get 2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=df1.columns.values, var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(10,5))
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df1_factors)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=4,init='k-means++',random_state=0).fit(df1_factors)
df1['cluster_num'] = kmeans.labels_
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
from collections import Counter
Counter(kmeans.labels_)
```

```python
sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="*", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

# Analyzing Start up Funding

```python
df2 = pd.read_csv('startup_funding(1).csv')
df2
```

```python
df2['SubVertical'].unique()
```

```python
df2=df2.drop(['Sr No','Date dd/mm/yyyy','Startup Name','Investors Name','InvestmentnType','Remarks'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2["Amount in USD"] = df2['Amount in USD'].str.replace(',','').str.extract('(^\d*)')
df2["Amount in USD"] = df2['Amount in USD'].replace('',np.nan)
df2['Amount in USD'] = df2['Amount in USD'].astype(float)
df2['Amount in USD'].fillna(df2['Amount in USD'].mean(),inplace=True)
```

```python
df2['SubVertical'].value_counts().head(10).plot.barh()
```

From above plot we can see that Online learning platform and online education platform are significant in numbers. Thus we can conclude that ed-tech market is a growing industry and can also evolve futhur.

```python
df2['Industry Vertical'] = df2['Industry Vertical'].apply(lambda x: 'ECommerce' if x=='eCommerce' else x)

new_df = df2.groupby('Industry Vertical')['Amount in USD'].sum().reset_index()
new_df.sort_values('Amount in USD', inplace=True, ascending=False)

new_df.set_index('Industry Vertical', inplace=True, drop=True)
new_df[:20].plot(kind='bar', figsize=(10,5))
plt.show()

new_df[:20].plot(kind='pie', subplots=True, figsize=(15,10), autopct='%.f%%')
plt.show()

new_df[:10].T
```

From above plot we can conclude that funding in ed tech start ups is currenlty very low but as we saw above that online education industry is a growing industry and so we can expect an increase in funding in future. 

# Customizing Marketing Mix

```python
df3 = pd.read_csv('Advertising.csv')
df3.head()
```

```python
corr_df = df3.corr()
corr_df
```

```python
plt.figure(figsize=(9,8))
sns.heatmap(round(df3.corr(),2),annot = True,cmap = 'YlGnBu',linewidths=3, fmt='.2g')
plt.title('Correlation Matrix')
plt.show()
```

```python
X = df3.loc[:, df3.columns != 'sales']
y = df3['sales']# Building Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)
model = RandomForestRegressor(random_state=1)
model.fit(X_train, y_train)
pred = model.predict(X_test)# Visualizing Feature Importance
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```

```python
import statsmodels.formula.api as sm
model = sm.ols(formula="sales~TV+radio+newspaper", data=df3).fit()
print(model.summary())
```

```python
from matplotlib.pyplot import figure

y_pred = model.predict()
labels = df3['sales']
df_temp = pd.DataFrame({'Actual': labels, 'Predicted':y_pred})
df_temp.head()

figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
y1 = df_temp['Actual']
y2 = df_temp['Predicted']
plt.plot(y1, label = 'Actual')
plt.plot(y2, label = 'Predicted')
plt.legend()
plt.show()
```
